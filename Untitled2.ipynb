```python
# Imports padronizados no in√≠cio
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
from tabulate import tabulate  # Para tabelas formatadas no terminal

# Fun√ß√£o para exibir tabelas formatadas com tabulate
def exibir_tabela(df, titulo=None, floatfmt=".2f", showindex=False):
    if titulo:
        print(titulo)
    print(tabulate(df, headers='keys', tablefmt='fancy_grid', showindex=showindex, floatfmt=floatfmt))
    print()

# Fun√ß√£o para gr√°fico de barras matplotlib com valores em cima
def grafico_barras_valores(x_labels, valores, titulo, xlabel, ylabel, cor, fmt='int', rotation=0):
    plt.figure(figsize=(10, 6))
    barras = plt.bar(x_labels, valores, color=cor, edgecolor='black', alpha=0.9)

    plt.title(titulo, fontsize=15, fontweight='bold')
    plt.ylabel(ylabel, fontsize=13)
    plt.xlabel(xlabel, fontsize=13)
    plt.xticks(rotation=rotation, fontsize=12)
    plt.yticks(fontsize=12)
    plt.grid(axis='y', linestyle='--', alpha=0.7)

    for barra in barras:
        altura = barra.get_height()
        label = f'{int(altura):,}' if fmt == 'int' else f'{altura:,.2f}'
        plt.text(
            barra.get_x() + barra.get_width() / 2,
            altura + max(valores) * 0.01,
            label.replace(',', '.'),
            ha='center',
            va='bottom',
            fontsize=11,
            fontweight='bold'
        )

    plt.tight_layout()
    plt.show()

# Fun√ß√£o para calcular estat√≠sticas b√°sicas agrupadas
def estatisticas_agrupadas(df, grupo, coluna):
    agrup = df.groupby(grupo)[coluna].agg(['sum', 'mean', 'std']).reset_index()
    agrup.columns = [grupo, f'Soma_{coluna}', f'M√©dia_{coluna}', f'Desvio_Padr√£o_{coluna}']
    return agrup

# Fun√ß√£o para sumarizar estat√≠sticas descritivas de m√∫ltiplas colunas
def resumo_descritivo(df, colunas, nomes):
    totais = df[colunas].sum().values
    medias = df[colunas].mean().values
    minimos = df[colunas].min().values
    maximos = df[colunas].max().values

    resumo = pd.DataFrame({
        'Total de Matr√≠culas': totais,
        'M√©dia (todos os registros)': medias,
        'M√≠nimo': minimos,
        'M√°ximo': maximos
    }, index=nomes)
    linha_total = {
        'Total de Matr√≠culas': totais.sum(),
        'M√©dia (todos os registros)': '-',
        'M√≠nimo': '-',
        'M√°ximo': '-'
    }
    resumo.loc['TOTAL GERAL'] = linha_total
    return resumo

# Carregamento e prepara√ß√£o dos dados
dataF = pd.read_csv('/content/MICRODADOS_CADASTRO_CURSOS_2023.CSV', encoding='latin1', sep=';')

print(f"Quantidade de linhas do dataset): {dataF.shape[0]}")
print(f"Quantidade de colunas do dataset): {dataF.shape[1]}\n")

# Classifica√ß√£o das vari√°veis em quantitativas e qualitativas
variaveis_quantitativas = [col for col in dataF.columns if pd.api.types.is_numeric_dtype(dataF[col])]
variaveis_qualitativas = [col for col in dataF.columns if dataF[col].dtype == 'object']

print("üìä Vari√°veis Quantitativas:")
for var in variaveis_quantitativas:
    print(f"- {var}")
print("\nüè∑Ô∏è Vari√°veis Qualitativas:")
for var in variaveis_qualitativas:
    print(f"- {var}")
print()

# Tabela de classifica√ß√£o das vari√°veis
classificacao = [{'Vari√°vel': col, 'Tipo': 'Qualitativa' if dataF[col].dtype == 'object' else 'Quantitativa'} for col in dataF.columns]
tabela_tipos = pd.DataFrame(classificacao)
exibir_tabela(tabela_tipos, "Classifica√ß√£o das Vari√°veis:\n", showindex=False)

# ===== Quest√£o 1 =====
print('Quest√£o 01 - Qual √© a distribui√ß√£o do n√∫mero de ingressantes por regi√£o do pa√≠s?')
print('Agrupamento por NO_REGIAO e c√°lculo da soma, m√©dia e desvio padr√£o da vari√°vel QT_ING.\n')

agrupado = estatisticas_agrupadas(dataF, 'NO_REGIAO', 'QT_ING')
exibir_tabela(agrupado, "üìã Tabela com Soma, M√©dia e Desvio Padr√£o de QT_ING por Regi√£o:\n")

print('GR√ÅFICOS DE BARRAS COM VALORES DA SOMA, M√âDIA E DESVIO PADR√ÉO POR REGI√ÉO\n')

grafico_barras_valores(agrupado['NO_REGIAO'], agrupado['Soma_QT_ING'],
                       'Soma de QT_ING por Regi√£o', 'Regi√£o', 'Soma', 'steelblue')

grafico_barras_valores(agrupado['NO_REGIAO'], agrupado['M√©dia_QT_ING'],
                       'M√©dia de QT_ING por Regi√£o', 'Regi√£o', 'M√©dia', 'seagreen', fmt='float')

grafico_barras_valores(agrupado['NO_REGIAO'], agrupado['Desvio_Padr√£o_QT_ING'],
                       'Desvio Padr√£o de QT_ING por Regi√£o', 'Regi√£o', 'Desvio Padr√£o', 'indianred', fmt='float')

print("\n\n")

# ===== Quest√£o 2 =====
print('Quest√£o 02 - Como √© distribu√≠do o n√∫mero de matr√≠culas por modalidade de ensino (presencial vs EAD)?')
print('Compara√ß√£o das vari√°veis QT_MAT_DIURNO, QT_MAT_NOTURNO e QT_MAT_EAD com medidas descritivas.\n')

# Calcular QT_MAT_EAD
dataF['QT_MAT_EAD'] = dataF['QT_MAT'] - (dataF['QT_MAT_DIURNO'] + dataF['QT_MAT_NOTURNO'])
dataF['QT_MAT_EAD'] = dataF['QT_MAT_EAD'].clip(lower=0)

# Definir colunas e nomes das modalidades
modalidades = ['QT_MAT_DIURNO', 'QT_MAT_NOTURNO', 'QT_MAT_EAD']
nomes_modalidade = ['Presencial Diurno', 'Presencial Noturno', 'EAD']

# Calcular medidas
totais = dataF[modalidades].sum().values
medias = dataF[modalidades].mean().values
minimos = dataF[modalidades].min().values
maximos = dataF[modalidades].max().values

# Criar DataFrame com a coluna adicional "Modalidade de Ensino"
tabela_modalidades = pd.DataFrame({
    'Modalidade de Ensino': nomes_modalidade,
    'Total de Matr√≠culas': totais,
    'M√©dia (todos os registros)': medias,
    'M√≠nimo': minimos,
    'M√°ximo': maximos
})

# Adicionar linha TOTAL GERAL
linha_total = pd.DataFrame({
    'Modalidade de Ensino': ['TOTAL GERAL'],
    'Total de Matr√≠culas': [totais.sum()],
    'M√©dia (todos os registros)': ['-'],
    'M√≠nimo': ['-'],
    'M√°ximo': ['-']
})

tabela_modalidades = pd.concat([tabela_modalidades, linha_total], ignore_index=True)

# Formatar tabela
tabela_formatada = tabela_modalidades.copy()
tabela_formatada['Total de Matr√≠culas'] = tabela_formatada['Total de Matr√≠culas'].apply(lambda x: f'{int(x):,}' if isinstance(x, (int, float)) else x)
tabela_formatada['M√©dia (todos os registros)'] = tabela_formatada['M√©dia (todos os registros)'].apply(lambda x: f'{x:,.2f}'.replace(',', 'X').replace('.', ',').replace('X', '.') if isinstance(x, (int, float)) else x)
tabela_formatada['M√≠nimo'] = tabela_formatada['M√≠nimo'].apply(lambda x: int(x) if isinstance(x, (int, float)) else x)
tabela_formatada['M√°ximo'] = tabela_formatada['M√°ximo'].apply(lambda x: int(x) if isinstance(x, (int, float)) else x)

# Exibir a tabela com coluna de Modalidade
exibir_tabela(tabela_formatada, "üìã Tabela com Matr√≠culas por Modalidade de Ensino:\n", showindex=False)

# Remover TOTAL GERAL para gr√°ficos
tabela_graf = tabela_modalidades[tabela_modalidades['Modalidade de Ensino'] != 'TOTAL GERAL']

# Gr√°ficos
grafico_barras_valores(tabela_graf['Modalidade de Ensino'], tabela_graf['Total de Matr√≠culas'],
                       'Total de Matr√≠culas por Modalidade de Ensino', 'Modalidade de Ensino', 'N√∫mero Total de Matr√≠culas', '#4C72B0')

grafico_barras_valores(tabela_graf['Modalidade de Ensino'], tabela_graf['M√©dia (todos os registros)'],
                       'M√©dia de Matr√≠culas por Modalidade (Incluindo Registros com 0)', 'Modalidade de Ensino', 'M√©dia por Registro', '#55A868', fmt='float')

grafico_barras_valores(tabela_graf['Modalidade de Ensino'], tabela_graf['M√≠nimo'],
                       'Menor N√∫mero de Matr√≠culas por Modalidade', 'Modalidade de Ensino', 'M√≠nimo de Matr√≠culas', '#C44E52')

grafico_barras_valores(tabela_graf['Modalidade de Ensino'], tabela_graf['M√°ximo'],
                       'Maior N√∫mero de Matr√≠culas por Modalidade', 'Modalidade de Ensino', 'M√°ximo de Matr√≠culas', '#8172B3')

# ===== Quest√£o 3 =====
print('Quest√£o 03 - Qual √© o perfil racial dos concluintes no Brasil?')
print('C√°lculo das frequ√™ncias absolutas e relativas das vari√°veis QT_CONC_*, com adi√ß√£o da coluna de perfil racial.\n')

# Vari√°veis de interesse
variaveis_raca_cor = [
    'QT_CONC_BRANCA',
    'QT_CONC_PRETA',
    'QT_CONC_PARDA',
    'QT_CONC_AMARELA',
    'QT_CONC_INDIGENA',
    'QT_CONC_CORND'
]

# Mapeamento para nomes leg√≠veis
nomes_raca_cor = {
    'QT_CONC_BRANCA': 'Branca',
    'QT_CONC_PRETA': 'Preta',
    'QT_CONC_PARDA': 'Parda',
    'QT_CONC_AMARELA': 'Amarela',
    'QT_CONC_INDIGENA': 'Ind√≠gena',
    'QT_CONC_CORND': 'N√£o Declarada'
}

# C√°lculo das frequ√™ncias
frequencia_absoluta = dataF[variaveis_raca_cor].sum()
total_concluintes = frequencia_absoluta.sum()
frequencia_relativa = (frequencia_absoluta / total_concluintes) * 100

# Montagem da nova tabela com coluna "Perfil Racial"
tabela_frequencias = pd.DataFrame({
    'Perfil Racial': [nomes_raca_cor[var] for var in variaveis_raca_cor],
    'Frequ√™ncia Absoluta': frequencia_absoluta.values,
    'Frequ√™ncia Relativa (%)': frequencia_relativa.values
}, index=variaveis_raca_cor)

# Linha total
linha_total = pd.DataFrame({
    'Perfil Racial': ['TOTAL GERAL'],
    'Frequ√™ncia Absoluta': [total_concluintes],
    'Frequ√™ncia Relativa (%)': [100.0]
}, index=['TOTAL GERAL'])

tabela_frequencias = pd.concat([tabela_frequencias, linha_total])

# Formata√ß√£o para exibi√ß√£o
tabela_frequencias_formatada = tabela_frequencias.copy()
tabela_frequencias_formatada['Frequ√™ncia Absoluta'] = tabela_frequencias_formatada['Frequ√™ncia Absoluta'].apply(lambda x: f'{int(x):,}'.replace(',', '.'))
tabela_frequencias_formatada['Frequ√™ncia Relativa (%)'] = tabela_frequencias_formatada['Frequ√™ncia Relativa (%)'].apply(lambda x: f'{x:,.2f}'.replace(',', 'X').replace('.', ',').replace('X', '.'))

exibir_tabela(tabela_frequencias_formatada, "üìã Frequ√™ncia de Concluintes por Perfil Racial:\n", showindex=False)

# Gr√°ficos (sem altera√ß√µes)
cores = ['#4C72B0', '#55A868', '#C44E52', '#8172B3', '#CCB974', '#64B5CD']

# Gr√°fico de barras - Frequ√™ncia Absoluta
nomes_legiveis = [nomes_raca_cor[var] for var in frequencia_absoluta.index]
valores = frequencia_absoluta.values

grafico_barras_valores(nomes_legiveis, valores,
                       'Frequ√™ncia Absoluta de Concluintes por Ra√ßa/Cor',
                       'Perfil Racial', 'N√∫mero de Concluintes', '#4C72B0', rotation=30)

# Gr√°ficos combinados: pizza e barras de frequ√™ncia relativa
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
explode = [0.05] * len(frequencia_relativa)
patches, texts = plt.pie(
    frequencia_relativa.values,
    labels=None,
    startangle=140,
    colors=cores,
    explode=explode,
    shadow=True,
    wedgeprops={'edgecolor': 'black'}
)
legenda = [f'{nomes_raca_cor[var]}: {valor:.2f}%' for var, valor in zip(variaveis_raca_cor, frequencia_relativa.values)]
plt.legend(patches, legenda, loc='center left', bbox_to_anchor=(1, 0.5), fontsize=12, title='Ra√ßa/Cor', title_fontsize=13)
plt.title('Frequ√™ncia Relativa (%) de Concluintes por Ra√ßa/Cor', fontsize=16, fontweight='bold')

plt.subplot(1, 2, 2)
barras = plt.bar(frequencia_relativa.index, frequencia_relativa.values, color=cores, edgecolor='black', alpha=0.9)
plt.title('Frequ√™ncia Relativa (%) de Concluintes por Ra√ßa/Cor', fontsize=16, fontweight='bold')
plt.ylabel('Percentual (%)', fontsize=13)
plt.xlabel('Ra√ßa/Cor', fontsize=13)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

for barra in barras:
    altura = barra.get_height()
    plt.text(barra.get_x() + barra.get_width() / 2, altura + 0.5, f'{altura:.2f}%', ha='center', va='bottom', fontsize=11, fontweight='bold')

plt.tight_layout()
plt.show()

# ===== Quest√£o 4 =====
print('Quest√£o 04 - Qual a m√©dia de estudantes com defici√™ncia matriculados por curso?')
print('- Estat√≠sticas por curso utilizando QT_MAT_DEFICIENTE.')
print('- Apresentamos apenas as 10 primeiras linhas.\n')

dataF['QT_MAT_DEFICIENTE'] = pd.to_numeric(dataF['QT_MAT_DEFICIENTE'], errors='coerce')

estatisticas_def = dataF.groupby('NO_CURSO')['QT_MAT_DEFICIENTE'].agg(
    M√©dia='mean',
    Mediana='median',
    Desvio_Padr√£o='std',
    M√≠nimo='min',
    M√°ximo='max'
).reset_index()

estatisticas_def[['M√©dia', 'Mediana', 'Desvio_Padr√£o', 'M√≠nimo', 'M√°ximo']] = estatisticas_def[
    ['M√©dia', 'Mediana', 'Desvio_Padr√£o', 'M√≠nimo', 'M√°ximo']
].round(2)

exibir_tabela(estatisticas_def.head(10), "üìä Estat√≠sticas por curso (10 primeiros):\n", floatfmt=".2f", showindex=False)

# Fun√ß√£o para gr√°fico interativo com plotly dos cursos com maiores m√©dias
def gerar_grafico_barras_interativo(df, n=10):
    top_cursos = df.sort_values(by='M√©dia', ascending=False).head(n)
    fig = px.bar(
        top_cursos,
        x='M√©dia',
        y='NO_CURSO',
        orientation='h',
        title=f'Top {n} Cursos com Maiores M√©dias de Alunos com Defici√™ncia',
        labels={'NO_CURSO': 'Curso', 'M√©dia': 'M√©dia de Alunos com Defici√™ncia'},
        text='M√©dia'
    )
    fig.update_layout(yaxis={'categoryorder': 'total ascending'}, template='plotly_white')
    fig.show()

gerar_grafico_barras_interativo(estatisticas_def, 10)
```
