```python
#importar a biblioteca pandas
import pandas as pd
#Carregando o dataset para o ambiente de trabalho
dataF = pd.read_csv('/content/MICRODADOS_CADASTRO_CURSOS_2023.CSV', encoding='latin1', sep=';')

#mostrando as primeiras linhas do dataset
dataF.head(5)

#Mostrando o tamanho do dataset, mais precisamente, o n√∫mero de linhas e colunas.
print("Quantidade de linhas do dataset):", dataF.shape[0])
print("Quantidade de colunas do dataset):", dataF.shape[1])

# Listas das vari√°veis
variaveis_quantitativas = []
variaveis_qualitativas = []

# Classifica√ß√£o das vari√°veis
for coluna in dataF.columns:
    if pd.api.types.is_numeric_dtype(dataF[coluna]):
        variaveis_quantitativas.append(coluna)
    elif dataF[coluna].dtype == 'object':
        variaveis_qualitativas.append(coluna)

# Exibir os resultados
print("üìä Vari√°veis Quantitativas:")
for var in variaveis_quantitativas:
    print(f"- {var}")

print("\nüè∑Ô∏è Vari√°veis Qualitativas:")
for var in variaveis_qualitativas:
    print(f"- {var}")

classificacao = []

# percorrendo as colunas(vari√°veis) e classificando entre quantitativa ou qualitativa.
for colunas in dataF.columns:
    if dataF[colunas].dtype == 'object':
        tipo = 'Qualitativa'
    elif pd.api.types.is_numeric_dtype(dataF[colunas]):
        tipo = 'Quantitativa'

    classificacao.append({'Vari√°vel': colunas, 'Tipo': tipo})

# criando tabela para organizar a classifica√ß√£o das vari√°veis.
tabela_tipos = pd.DataFrame(classificacao)


print(tabela_tipos)
print()
print()

#Quest√£o 1
print ('Quest√£o 01 - Qual √© a distribui√ß√£o do n√∫mero de ingressantes por regi√£o do pa√≠s?')
print('Agrupamento por NO_REGIAO e calculo da soma, m√©dia e desvio padr√£o da vari√°vel QT_ING.')



# Agrupar os dados
agrupado = dataF.groupby('NO_REGIAO')['QT_ING'].agg(['sum', 'mean', 'std']).reset_index()
agrupado.columns = ['Regi√£o', 'Soma_QT_ING', 'M√©dia_QT_ING', 'Desvio_Padr√£o_QT_ING']

# Tabela criada com os dados agrupados por Regi√£o
print("\nüìã Tabela com Soma, M√©dia e Desvio Padr√£o de QT_ING por Regi√£o:\n")
print(agrupado)
print()

print('GR√ÅFICOS DE BARRAS TRAZENDO UMA MELHOR VISUALIZA√á√ÉO DA SOMA, M√âDIA E DESVIO PADR√ÉO POR REGI√ÉO')
print()

def grafico_com_valores(titulo, y_label, valores, cor, label_legenda):
    plt.figure(figsize=(9, 5))
    barras = plt.bar(agrupado['Regi√£o'], valores, color=cor, label=label_legenda)
    plt.title(titulo, fontsize=14)
    plt.xlabel('Regi√£o', fontsize=12)
    plt.ylabel(y_label, fontsize=12)
    plt.xticks(rotation=45)
    plt.legend(loc='upper right', fontsize=10)

    for barra in barras:
        altura = barra.get_height()
        plt.text(
            barra.get_x() + barra.get_width() / 2,
            altura,
            f'{altura:.2f}',                       # valor com 2 casas decimais
            ha='center', va='bottom', fontsize=9, color='black'
        )

    plt.tight_layout()
    plt.show()

# Gr√°ficos com valores em cima das barras e legenda
grafico_com_valores('Soma de QT_ING por Regi√£o', 'Soma', agrupado['Soma_QT_ING'], 'steelblue', 'Soma de QT_ING')
grafico_com_valores('M√©dia de QT_ING por Regi√£o', 'M√©dia', agrupado['M√©dia_QT_ING'], 'seagreen', 'M√©dia de QT_ING')
grafico_com_valores('Desvio Padr√£o de QT_ING por Regi√£o', 'Desvio Padr√£o', agrupado['Desvio_Padr√£o_QT_ING'], 'indianred', 'Desvio Padr√£o de QT_ING')

print()
print()
print()

#Quest√£o 2
print ('Quest√£o 02 - Como √© distribu√≠do o n√∫mero de matr√≠culas por modalidade de ensino (presencial vs EAD)?')
print('Compara√ß√£o da QT_MAT_DIURNO, QT_MAT_NOTURNO e QT_MAT_EAD atrav√©s de medidas descritivas (m√©dia, mediana, m√≠nimo, m√°ximo).')

print()
print()
import pandas as pd
import matplotlib.pyplot as plt

# Calcular QT_MAT_EAD
dataF['QT_MAT_EAD'] = dataF['QT_MAT'] - (dataF['QT_MAT_DIURNO'] + dataF['QT_MAT_NOTURNO'])
dataF['QT_MAT_EAD'] = dataF['QT_MAT_EAD'].clip(lower=0)

# Definir colunas e nomes das modalidades
modalidades = ['QT_MAT_DIURNO', 'QT_MAT_NOTURNO', 'QT_MAT_EAD']
nomes_modalidade = ['Presencial Diurno', 'Presencial Noturno', 'EAD']

# Calcular medidas
totais = dataF[modalidades].sum().values
medias = dataF[modalidades].mean().values  # Inclui registros com 0
minimos = dataF[modalidades].min().values
maximos = dataF[modalidades].max().values

# Criar tabela final
tabela = pd.DataFrame({
    'Total de Matr√≠culas': totais,
    'M√©dia (todos os registros)': medias,
    'M√≠nimo': minimos,
    'M√°ximo': maximos
}, index=nomes_modalidade)

# Adicionar linha TOTAL GERAL
linha_total = {
    'Total de Matr√≠culas': totais.sum(),
    'M√©dia (todos os registros)': '-',
    'M√≠nimo': '-',
    'M√°ximo': '-'
}
tabela.loc['TOTAL GERAL'] = linha_total

# Exibir tabela
print("üìã Tabela com Total, M√©dia (todos os registros), M√≠nimo e M√°ximo:\n")
print(tabela)

# Remover TOTAL GERAL para gr√°ficos
tabela_graf = tabela.drop('TOTAL GERAL')

# Fun√ß√£o gen√©rica para gr√°ficos (Total, M√©dia, M√≠nimo, M√°ximo)
def grafico_barras(coluna, titulo, ylabel, cor, fmt='int'):
    plt.figure(figsize=(10, 6))
    valores = tabela_graf[coluna]
    barras = plt.bar(tabela_graf.index, valores, color=cor, edgecolor='black', alpha=0.9)

    plt.title(titulo, fontsize=15, fontweight='bold')
    plt.ylabel(ylabel, fontsize=13)
    plt.xlabel('Modalidade de Ensino', fontsize=13)
    plt.xticks(rotation=0, fontsize=12)
    plt.yticks(fontsize=12)
    plt.grid(axis='y', linestyle='--', alpha=0.7)

    # Adicionar valores em cima das barras
    for barra in barras:
        altura = barra.get_height()
        label = f'{int(altura):,}' if fmt == 'int' else f'{altura:,.2f}'
        plt.text(
            barra.get_x() + barra.get_width() / 2,
            altura + max(valores) * 0.01,
            label,
            ha='center',
            va='bottom',
            fontsize=11,
            fontweight='bold'
        )

    plt.tight_layout()
    plt.show()

# Gr√°ficos para cada medida
grafico_barras('Total de Matr√≠culas', 'Total de Matr√≠culas por Modalidade de Ensino', 'N√∫mero Total de Matr√≠culas', '#4C72B0')
grafico_barras('M√©dia (todos os registros)', 'M√©dia de Matr√≠culas por Modalidade (Incluindo Registros com 0)', 'M√©dia por Registro', '#55A868', fmt='float')
grafico_barras('M√≠nimo', 'Menor N√∫mero de Matr√≠culas por Modalidade', 'M√≠nimo de Matr√≠culas', '#C44E52')
grafico_barras('M√°ximo', 'Maior N√∫mero de Matr√≠culas por Modalidade', 'M√°ximo de Matr√≠culas', '#8172B3')
print()
print()
print()
#Quest√£o 3
print ('Quest√£o 03 - Qual √© o perfil racial dos concluintes no Brasil?')
print('Calculo das frequ√™ncias absolutas e relativas √†s vari√°veis ‚Äã‚ÄãQT_CONC_BRANCA, QT_CONC_PRETA, QT_CONC_PARDA, QT_CONC_AMARELA, QT_CONC_INDIGENA e QT_CONC_CORND.')

print()
print()
# Vari√°veis de interesse
variaveis_raca_cor = [
    'QT_CONC_BRANCA',
    'QT_CONC_PRETA',
    'QT_CONC_PARDA',
    'QT_CONC_AMARELA',
    'QT_CONC_INDIGENA',
    'QT_CONC_CORND'
]

# Calcular frequ√™ncia absoluta (total de concluintes por cor/ra√ßa)
frequencia_absoluta = dataF[variaveis_raca_cor].sum()

# Calcular frequ√™ncia relativa (percentual sobre o total geral de concluintes)
total_concluintes = frequencia_absoluta.sum()
frequencia_relativa = (frequencia_absoluta / total_concluintes) * 100

# Montar tabela com ambas as frequ√™ncias
tabela_frequencias = pd.DataFrame({
    'Frequ√™ncia Absoluta': frequencia_absoluta,
    'Frequ√™ncia Relativa (%)': frequencia_relativa.round(2)
})

# Adicionar linha TOTAL GERAL
linha_total = pd.DataFrame({
    'Frequ√™ncia Absoluta': [total_concluintes],
    'Frequ√™ncia Relativa (%)': [100.00]
}, index=['TOTAL GERAL'])

tabela_frequencias = pd.concat([tabela_frequencias, linha_total])

# Formata√ß√£o dos valores
tabela_frequencias['Frequ√™ncia Absoluta'] = tabela_frequencias['Frequ√™ncia Absoluta'].apply(lambda x: f'{int(x):,}'.replace(',', '.'))
tabela_frequencias['Frequ√™ncia Relativa (%)'] = tabela_frequencias['Frequ√™ncia Relativa (%)'].apply(lambda x: f'{x:,.2f}'.replace(',', 'X').replace('.', ',').replace('X', '.'))

print("Frequ√™ncias Absolutas e Relativas das vari√°veis por ra√ßa/cor:\n")
print(tabela_frequencias)

import matplotlib.pyplot as plt

# Dados (reutilizando a tabela formatada anteriormente, mas com valores num√©ricos para os gr√°ficos)
# Recriar dados num√©ricos para evitar conflito com strings formatadas
freq_abs = frequencia_absoluta
freq_rel = frequencia_relativa

# Remover o total geral para os gr√°ficos
freq_abs_graf = freq_abs.copy()
freq_rel_graf = freq_rel.copy()

# Gr√°fico de barras - Frequ√™ncia Absoluta
plt.figure(figsize=(10, 6))
barras = plt.bar(freq_abs_graf.index, freq_abs_graf.values, color='#4C72B0', edgecolor='black', alpha=0.85)

plt.title('Frequ√™ncia Absoluta de Concluintes por Ra√ßa/Cor', fontsize=16, fontweight='bold')
plt.ylabel('N√∫mero de Concluintes', fontsize=14)
plt.xlabel('Ra√ßa/Cor', fontsize=14)
plt.xticks(rotation=30, fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Valores em cima das barras
for barra in barras:
    altura = barra.get_height()
    plt.text(
        barra.get_x() + barra.get_width() / 2,
        altura + max(freq_abs_graf.values)*0.01,
        f'{int(altura):,}'.replace(',', '.'),
        ha='center',
        va='bottom',
        fontsize=11,
        fontweight='bold'
    )

plt.tight_layout()
plt.show()

freq_abs = frequencia_absoluta
freq_rel = frequencia_relativa
cores = ['#4C72B0', '#55A868', '#C44E52', '#8172B3', '#CCB974', '#64B5CD']

# Gr√°fico de pizza corrigido
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
explode = [0.05]*len(freq_rel)
patches, texts = plt.pie(
    freq_rel.values,
    labels=None,
    startangle=140,
    colors=cores,
    explode=explode,
    shadow=True,
    wedgeprops={'edgecolor': 'black'}
)
legenda = [f'{cat}: {valor:.2f}%' for cat, valor in zip(freq_rel.index, freq_rel.values)]
plt.legend(patches, legenda, loc='center left', bbox_to_anchor=(1, 0.5),
           fontsize=12, title='Ra√ßa/Cor', title_fontsize=13)
plt.title('Frequ√™ncia Relativa (%) de Concluintes por Ra√ßa/Cor', fontsize=16, fontweight='bold')

# Gr√°fico de barras para frequ√™ncia relativa
plt.subplot(1, 2, 2)
barras = plt.bar(freq_rel.index, freq_rel.values, color=cores, edgecolor='black', alpha=0.9)
plt.title('Frequ√™ncia Relativa (%) de Concluintes por Ra√ßa/Cor', fontsize=16, fontweight='bold')
plt.ylabel('Percentual (%)', fontsize=13)
plt.xlabel('Ra√ßa/Cor', fontsize=13)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Valores em cima das barras com duas casas decimais
for barra in barras:
    altura = barra.get_height()
    plt.text(
        barra.get_x() + barra.get_width() / 2,
        altura + 0.5,
        f'{altura:.2f}%',
        ha='center',
        va='bottom',
        fontsize=11,
        fontweight='bold'
    )

plt.tight_layout()
plt.show()
```
