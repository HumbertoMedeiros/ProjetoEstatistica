```python
# Imports padronizados no início
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
from tabulate import tabulate  # Para tabelas formatadas no terminal

# Função para exibir tabelas formatadas com tabulate
def exibir_tabela(df, titulo=None, floatfmt=".2f", showindex=False):
    if titulo:
        print(titulo)
    print(tabulate(df, headers='keys', tablefmt='fancy_grid', showindex=showindex, floatfmt=floatfmt))
    print()

# Função para gráfico de barras matplotlib com valores em cima
def grafico_barras_valores(x, y, titulo, xlabel, ylabel, cor, rotacao=0, fmt='int'):
    plt.figure(figsize=(10, 6))
    barras = plt.bar(x, y, color=cor, edgecolor='black', alpha=0.9)
    plt.title(titulo, fontsize=15, fontweight='bold')
    plt.xlabel(xlabel, fontsize=13)
    plt.ylabel(ylabel, fontsize=13)
    plt.xticks(rotation=rotacao, fontsize=12)
    plt.yticks(fontsize=12)
    plt.grid(axis='y', linestyle='--', alpha=0.7)

    for barra in barras:
        altura = barra.get_height()
        label = f'{int(altura):,}'.replace(',', '.') if fmt == 'int' else f'{altura:,.2f}'.replace(',', 'X').replace('.', ',').replace('X', '.')
        plt.text(barra.get_x() + barra.get_width() / 2, altura + max(y)*0.01, label, ha='center', va='bottom', fontsize=11, fontweight='bold')

    plt.tight_layout()
    plt.show()

# Função para calcular estatísticas básicas agrupadas
def estatisticas_agrupadas(df, grupo, coluna):
    agrup = df.groupby(grupo)[coluna].agg(['sum', 'mean', 'std']).reset_index()
    agrup.columns = [grupo, f'Soma_{coluna}', f'Média_{coluna}', f'Desvio_Padrão_{coluna}']
    return agrup

# Função para sumarizar estatísticas descritivas de múltiplas colunas
def resumo_descritivo(df, colunas, nomes):
    totais = df[colunas].sum().values
    medias = df[colunas].mean().values
    minimos = df[colunas].min().values
    maximos = df[colunas].max().values

    resumo = pd.DataFrame({
        'Total de Matrículas': totais,
        'Média (todos os registros)': medias,
        'Mínimo': minimos,
        'Máximo': maximos
    }, index=nomes)
    linha_total = {
        'Total de Matrículas': totais.sum(),
        'Média (todos os registros)': '-',
        'Mínimo': '-',
        'Máximo': '-'
    }
    resumo.loc['TOTAL GERAL'] = linha_total
    return resumo

# Carregamento e preparação dos dados
dataF = pd.read_csv('/content/MICRODADOS_CADASTRO_CURSOS_2023.CSV', encoding='latin1', sep=';')

print(f"Quantidade de linhas do dataset): {dataF.shape[0]}")
print(f"Quantidade de colunas do dataset): {dataF.shape[1]}\n")

# Classificação das variáveis em quantitativas e qualitativas
variaveis_quantitativas = [col for col in dataF.columns if pd.api.types.is_numeric_dtype(dataF[col])]
variaveis_qualitativas = [col for col in dataF.columns if dataF[col].dtype == 'object']

print("📊 Variáveis Quantitativas:")
for var in variaveis_quantitativas:
    print(f"- {var}")
print("\n🏷️ Variáveis Qualitativas:")
for var in variaveis_qualitativas:
    print(f"- {var}")
print()

# Tabela de classificação das variáveis
classificacao = [{'Variável': col, 'Tipo': 'Qualitativa' if dataF[col].dtype == 'object' else 'Quantitativa'} for col in dataF.columns]
tabela_tipos = pd.DataFrame(classificacao)
exibir_tabela(tabela_tipos, "Classificação das Variáveis:\n", showindex=False)

# ===== Questão 1 =====
print('Questão 01 - Qual é a distribuição do número de ingressantes por região do país?')
print('Agrupamento por NO_REGIAO e cálculo da soma, média e desvio padrão da variável QT_ING.\n')

agrupado = estatisticas_agrupadas(dataF, 'NO_REGIAO', 'QT_ING')
exibir_tabela(agrupado, "📋 Tabela com Soma, Média e Desvio Padrão de QT_ING por Região:\n")

print('GRÁFICOS DE BARRAS COM VALORES DA SOMA, MÉDIA E DESVIO PADRÃO POR REGIÃO\n')

grafico_barras_valores(agrupado['NO_REGIAO'], agrupado['Soma_QT_ING'],
                       'Soma de QT_ING por Região', 'Região', 'Soma', 'steelblue')

grafico_barras_valores(agrupado['NO_REGIAO'], agrupado['Média_QT_ING'],
                       'Média de QT_ING por Região', 'Região', 'Média', 'seagreen', fmt='float')

grafico_barras_valores(agrupado['NO_REGIAO'], agrupado['Desvio_Padrão_QT_ING'],
                       'Desvio Padrão de QT_ING por Região', 'Região', 'Desvio Padrão', 'indianred', fmt='float')

print("\n\n")

# ===== Questão 2 =====
print('Questão 02 - Como é distribuído o número de matrículas por modalidade de ensino (presencial vs EAD)?')
print('Comparação das variáveis QT_MAT_DIURNO, QT_MAT_NOTURNO e QT_MAT_EAD com medidas descritivas.\n')

# Calcular QT_MAT_EAD
dataF['QT_MAT_EAD'] = dataF['QT_MAT'] - (dataF['QT_MAT_DIURNO'] + dataF['QT_MAT_NOTURNO'])
dataF['QT_MAT_EAD'] = dataF['QT_MAT_EAD'].clip(lower=0)

# Definir colunas e nomes das modalidades
modalidades = ['QT_MAT_DIURNO', 'QT_MAT_NOTURNO', 'QT_MAT_EAD']
nomes_modalidade = ['Presencial Diurno', 'Presencial Noturno', 'EAD']

# Calcular medidas
totais = dataF[modalidades].sum().values
medias = dataF[modalidades].mean().values
minimos = dataF[modalidades].min().values
maximos = dataF[modalidades].max().values

# Criar DataFrame com a coluna adicional "Modalidade de Ensino"
tabela_modalidades = pd.DataFrame({
    'Modalidade de Ensino': nomes_modalidade,
    'Total de Matrículas': totais,
    'Média (todos os registros)': medias,
    'Mínimo': minimos,
    'Máximo': maximos
})

# Adicionar linha TOTAL GERAL
linha_total = pd.DataFrame({
    'Modalidade de Ensino': ['TOTAL GERAL'],
    'Total de Matrículas': [totais.sum()],
    'Média (todos os registros)': ['-'],
    'Mínimo': ['-'],
    'Máximo': ['-']
})

tabela_modalidades = pd.concat([tabela_modalidades, linha_total], ignore_index=True)

# Formatar tabela
tabela_formatada = tabela_modalidades.copy()
tabela_formatada['Total de Matrículas'] = tabela_formatada['Total de Matrículas'].apply(lambda x: f'{int(x):,}' if isinstance(x, (int, float)) else x)
tabela_formatada['Média (todos os registros)'] = tabela_formatada['Média (todos os registros)'].apply(lambda x: f'{x:,.2f}'.replace(',', 'X').replace('.', ',').replace('X', '.') if isinstance(x, (int, float)) else x)
tabela_formatada['Mínimo'] = tabela_formatada['Mínimo'].apply(lambda x: int(x) if isinstance(x, (int, float)) else x)
tabela_formatada['Máximo'] = tabela_formatada['Máximo'].apply(lambda x: int(x) if isinstance(x, (int, float)) else x)

# Exibir a tabela com coluna de Modalidade
exibir_tabela(tabela_formatada, "📋 Tabela com Matrículas por Modalidade de Ensino:\n", showindex=False)

# Remover TOTAL GERAL para gráficos
tabela_graf = tabela_modalidades[tabela_modalidades['Modalidade de Ensino'] != 'TOTAL GERAL']

# Gráficos
grafico_barras_valores(tabela_graf['Modalidade de Ensino'], tabela_graf['Total de Matrículas'],
                       'Total de Matrículas por Modalidade de Ensino', 'Modalidade de Ensino', 'Número Total de Matrículas', '#4C72B0')

grafico_barras_valores(tabela_graf['Modalidade de Ensino'], tabela_graf['Média (todos os registros)'],
                       'Média de Matrículas por Modalidade (Incluindo Registros com 0)', 'Modalidade de Ensino', 'Média por Registro', '#55A868', fmt='float')

grafico_barras_valores(tabela_graf['Modalidade de Ensino'], tabela_graf['Mínimo'],
                       'Menor Número de Matrículas por Modalidade', 'Modalidade de Ensino', 'Mínimo de Matrículas', '#C44E52')

grafico_barras_valores(tabela_graf['Modalidade de Ensino'], tabela_graf['Máximo'],
                       'Maior Número de Matrículas por Modalidade', 'Modalidade de Ensino', 'Máximo de Matrículas', '#8172B3')

# ===== Questão 3 =====
print('Questão 03 - Qual é o perfil racial dos concluintes no Brasil?')
print('Cálculo das frequências absolutas e relativas das variáveis QT_CONC_*, com adição da coluna de perfil racial.\n')

# Variáveis de interesse
variaveis_raca_cor = [
    'QT_CONC_BRANCA',
    'QT_CONC_PRETA',
    'QT_CONC_PARDA',
    'QT_CONC_AMARELA',
    'QT_CONC_INDIGENA',
    'QT_CONC_CORND'
]

# Mapeamento para nomes legíveis
nomes_raca_cor = {
    'QT_CONC_BRANCA': 'Branca',
    'QT_CONC_PRETA': 'Preta',
    'QT_CONC_PARDA': 'Parda',
    'QT_CONC_AMARELA': 'Amarela',
    'QT_CONC_INDIGENA': 'Indígena',
    'QT_CONC_CORND': 'Não Declarada'
}

# Cálculo das frequências
frequencia_absoluta = dataF[variaveis_raca_cor].sum()
total_concluintes = frequencia_absoluta.sum()
frequencia_relativa = (frequencia_absoluta / total_concluintes) * 100

# Montagem da nova tabela com coluna "Perfil Racial"
tabela_frequencias = pd.DataFrame({
    'Perfil Racial': [nomes_raca_cor[var] for var in variaveis_raca_cor],
    'Frequência Absoluta': frequencia_absoluta.values,
    'Frequência Relativa (%)': frequencia_relativa.values
}, index=variaveis_raca_cor)

# Linha total
linha_total = pd.DataFrame({
    'Perfil Racial': ['TOTAL GERAL'],
    'Frequência Absoluta': [total_concluintes],
    'Frequência Relativa (%)': [100.0]
}, index=['TOTAL GERAL'])

tabela_frequencias = pd.concat([tabela_frequencias, linha_total])

# Formatação para exibição
tabela_frequencias_formatada = tabela_frequencias.copy()
tabela_frequencias_formatada['Frequência Absoluta'] = tabela_frequencias_formatada['Frequência Absoluta'].apply(lambda x: f'{int(x):,}'.replace(',', '.'))
tabela_frequencias_formatada['Frequência Relativa (%)'] = tabela_frequencias_formatada['Frequência Relativa (%)'].apply(lambda x: f'{x:,.2f}'.replace(',', 'X').replace('.', ',').replace('X', '.'))

exibir_tabela(tabela_frequencias_formatada, "📋 Frequência de Concluintes por Perfil Racial:\n", showindex=False)

# Gráficos (sem alterações)
cores = ['#4C72B0', '#55A868', '#C44E52', '#8172B3', '#CCB974', '#64B5CD']

# Gráfico de barras - Frequência Absoluta
grafico_barras_valores(frequencia_absoluta.index, frequencia_absoluta.values,
                       'Frequência Absoluta de Concluintes por Raça/Cor', 'Raça/Cor', 'Número de Concluintes', '#4C72B0')

# Gráficos combinados: pizza e barras de frequência relativa
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
explode = [0.05] * len(frequencia_relativa)
patches, texts = plt.pie(
    frequencia_relativa.values,
    labels=None,
    startangle=140,
    colors=cores,
    explode=explode,
    shadow=True,
    wedgeprops={'edgecolor': 'black'}
)
legenda = [f'{nomes_raca_cor[var]}: {valor:.2f}%' for var, valor in zip(variaveis_raca_cor, frequencia_relativa.values)]
plt.legend(patches, legenda, loc='center left', bbox_to_anchor=(1, 0.5), fontsize=12, title='Raça/Cor', title_fontsize=13)
plt.title('Frequência Relativa (%) de Concluintes por Raça/Cor', fontsize=16, fontweight='bold')

plt.subplot(1, 2, 2)
barras = plt.bar(frequencia_relativa.index, frequencia_relativa.values, color=cores, edgecolor='black', alpha=0.9)
plt.title('Frequência Relativa (%) de Concluintes por Raça/Cor', fontsize=16, fontweight='bold')
plt.ylabel('Percentual (%)', fontsize=13)
plt.xlabel('Raça/Cor', fontsize=13)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

for barra in barras:
    altura = barra.get_height()
    plt.text(barra.get_x() + barra.get_width() / 2, altura + 0.5, f'{altura:.2f}%', ha='center', va='bottom', fontsize=11, fontweight='bold')

plt.tight_layout()
plt.show()

# ===== Questão 4 =====
print('Questão 04 - Qual a média de estudantes com deficiência matriculados por curso?')
print('- Estatísticas por curso utilizando QT_MAT_DEFICIENTE.')
print('- Apresentamos apenas as 10 primeiras linhas.\n')

dataF['QT_MAT_DEFICIENTE'] = pd.to_numeric(dataF['QT_MAT_DEFICIENTE'], errors='coerce')

estatisticas_def = dataF.groupby('NO_CURSO')['QT_MAT_DEFICIENTE'].agg(
    Média='mean',
    Mediana='median',
    Desvio_Padrão='std',
    Mínimo='min',
    Máximo='max'
).reset_index()

estatisticas_def[['Média', 'Mediana', 'Desvio_Padrão', 'Mínimo', 'Máximo']] = estatisticas_def[
    ['Média', 'Mediana', 'Desvio_Padrão', 'Mínimo', 'Máximo']
].round(2)

exibir_tabela(estatisticas_def.head(10), "📊 Estatísticas por curso (10 primeiros):\n", floatfmt=".2f", showindex=False)

# Função para gráfico interativo com plotly dos cursos com maiores médias
def gerar_grafico_barras_interativo(df, n=10):
    top_cursos = df.sort_values(by='Média', ascending=False).head(n)
    fig = px.bar(
        top_cursos,
        x='Média',
        y='NO_CURSO',
        orientation='h',
        title=f'Top {n} Cursos com Maiores Médias de Alunos com Deficiência',
        labels={'NO_CURSO': 'Curso', 'Média': 'Média de Alunos com Deficiência'},
        text='Média'
    )
    fig.update_layout(yaxis={'categoryorder': 'total ascending'}, template='plotly_white')
    fig.show()

gerar_grafico_barras_interativo(estatisticas_def, 10)
```
