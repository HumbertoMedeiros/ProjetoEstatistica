```python
#importar a biblioteca pandas
import pandas as pd
#Carregando o dataset para o ambiente de trabalho
dataF = pd.read_csv('/content/MICRODADOS_CADASTRO_CURSOS_2023.CSV', encoding='latin1', sep=';')

#mostrando as primeiras linhas do dataset
dataF.head(5)

#Mostrando o tamanho do dataset, mais precisamente, o n√∫mero de linhas e colunas.
print("Quantidade de linhas do dataset):", dataF.shape[0])
print("Quantidade de colunas do dataset):", dataF.shape[1])

# Listas das vari√°veis
variaveis_quantitativas = []
variaveis_qualitativas = []

# Classifica√ß√£o das vari√°veis
for coluna in dataF.columns:
    if pd.api.types.is_numeric_dtype(dataF[coluna]):
        variaveis_quantitativas.append(coluna)
    elif dataF[coluna].dtype == 'object':
        variaveis_qualitativas.append(coluna)

# Exibir os resultados
print("üìä Vari√°veis Quantitativas:")
for var in variaveis_quantitativas:
    print(f"- {var}")

print("\nüè∑Ô∏è Vari√°veis Qualitativas:")
for var in variaveis_qualitativas:
    print(f"- {var}")

classificacao = []

# percorrendo as colunas(vari√°veis) e classificando entre quantitativa ou qualitativa.
for colunas in dataF.columns:
    if dataF[colunas].dtype == 'object':
        tipo = 'Qualitativa'
    elif pd.api.types.is_numeric_dtype(dataF[colunas]):
        tipo = 'Quantitativa'

    classificacao.append({'Vari√°vel': colunas, 'Tipo': tipo})

# criando tabela para organizar a classifica√ß√£o das vari√°veis.
tabela_tipos = pd.DataFrame(classificacao)


print(tabela_tipos)


```
